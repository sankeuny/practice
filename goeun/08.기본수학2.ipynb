### 소수_ 에라토스테네스의 체 알고리즘

2부터 n까지 돌면서 각 수의 배수를 지워나가는 방식

```python
# 에라토스테네스의 체 - 소수를 찾고 싶은 경우 '그 수의 제곱근보다 작은 소수의 배수를 없애면 남은 수가 소수'
# n이하의 수중 소수의 개수 찾기

def solution(n):
    answer = 0   # 갯수 담을 변수
    number = []  # n이하의 수중 소수 파악

    for i in range(n+1):
        number.append(i)  # n까지의 모든 수를 number 리스트에 담기 [1,2,3,4,5, ... ,n]

    for i in range(2, n+1, 1):     # i = 2
        if i == -1:
            continue
        j = 0                      # i = 2 
        for j in range(i+i, n+1, i):  # 2 + 2 즉, 2의배수는 소수가 아니므로 지운다
            number[j] = -1   # 소수가 아닌 방에는 -1을 담기 -1은 소수가 아님을 표시

    for i in range(2, n+1):
        if number[i] != -1:
            answer += 1
            
    return answer

print(solution(10))
```

### 1978 소수찾기
> **문제** <br>
주어진 수 N개 중에서 소수가 몇 개인지 찾아서 출력하는 프로그램을 작성하시오. 



```python
a = int(input())
is_prime = list(map(int, input().split()))
prime = 0 # 소수 개수

# 에라테스테네스의 체  
for i in is_prime: 
    e = 0 
    if i > 1:
        for j in range(2, i):
            if  i % j == 0:
                e = 1   # 배수
        if e == 0:
            prime += 1 # 소수
                
print(prime)
```

### 2581 소수
> **문제** <br>
자연수 M과 N이 주어질 때 M이상 N이하의 자연수 중 소수인 것을 모두 골라 이들 소수의 합과 최솟값을 찾는 프로그램을 작성하시오.

예를 들어 M=60, N=100인 경우 60이상 100이하의 자연수 중 소수는 61, 67, 71, 73, 79, 83, 89, 97 총 8개가 있으므로, 이들 소수의 합은 620이고, 최솟값은 61이 된다.

```python
a = int(input())
b = int(input())
array = []
for i in range(a,b+1):
    e = 0
    if i > 1:
        for j in range(2,i):
            if i % j == 0:
                e = 1
        if e == 0:
            array.append(i)
            
if len(array) > 0:
    print(sum(array))
    print(min(array))
else:
    print(-1)
```

### 11653 소인수분해
> **문제** <br>
정수 N이 주어졌을 때, 소인수분해하는 프로그램을 작성하시오.

```python
# 소수 판별해서 식구하면 오래걸림 -> 소인수분해 알고리즘 참고
n = int(input())  # 나누어지는 수
d = 2  # 나누는 수

while n != 1: # 나누어 떨어지지 않으면
    if n % d != 0: # 나누는 수 1 증가
        d += 1 #  나누어 떨어지면
    else:
        print(d) # 소인수니까 출력하고
        n //= d  # 나누어지는 수도 갱신
```

```python
# 시간초과 -> 제곱하여 시간을 단축
def is_prime(num):
    if num == 1:
        return False
    else:
        # n의 최대 약수가 sqrt(n) 이하이므로 i=sqrt(n)까지 검사
        for i in range(2, int(num ** 0.5) + 1):  
            if num % i == 0:
                return False
        return True

x,y = map(int, input().split())
for i in range(x,y+1):
    if is_prime(i):
        print(i)
```

### 4948 베르트랑 공준
> **문제** <br>
베르트랑 공준은 임의의 자연수 n에 대하여, n보다 크고, 2n보다 작거나 같은 소수는 적어도 하나 존재한다는 내용을 담고 있다. <br>
이 명제는 조제프 베르트랑이 1845년에 추측했고, 파프누티 체비쇼프가 1850년에 증명했다.<br>
예를 들어, 10보다 크고, 20보다 작거나 같은 소수는 4개가 있다. (11, 13, 17, 19) 또, 14보다 크고, 28보다 작거나 같은 소수는 3개가 있다. (17,19, 23)<br>
자연수 n이 주어졌을 때, n보다 크고, 2n보다 작거나 같은 소수의 개수를 구하는 프로그램을 작성하시오. 

```python
# 시간 초과...
# 소수를 구하는 함수 -> while 문 
def is_prime(a):
    prime = 0  
    for i in range(a+1, (2*a)+1): 
        e = 0 
        if i > 1:
            for j in range(2, int(i**0.5)+1):
                if  i % j == 0:
                    e = 1  
            if e == 0:
                prime += 1
        elif i == 1:
            prime += 1
    print(prime)

while True:
    n = int(input())
    if n == 0:
        break
    is_prime(n)
```
